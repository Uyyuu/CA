```python
from sklearn.datasets import fetch_covtype
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
#data,target
d,t = fetch_covtype(return_X_y=True)
covtype = fetch_covtype()

#特徴量のスケーリング(標準化)
std = StandardScaler()
d_std = std.fit_transform(d)

#DataFrameに変換したもの
df = pd.DataFrame(covtype.data,columns=covtype.feature_names)
df.head()
```

```python
kf1 = KFold(n_splits=5, shuffle=True, random_state=100)
kf2 = KFold(n_splits=3, shuffle=True, random_state=200)
kf3 = KFold(n_splits=5, shuffle=True, random_state=300)
```

決定木

```python
dtc1 = DecisionTreeClassifier(max_depth=5)
scores_dtc1 = cross_val_score(dtc1, d, t, cv=5)
print("Avr-Score-by-dtc1:",scores_dtc1.mean())
print("Std-Score-by-dtc1:",scores_dtc1.std())

dtc1.fit(d,t)
importances_dtc1 = dtc1.feature_importances_
```

```python
dtc_gscv = DecisionTreeClassifier(random_state=0)

param_grid = {'max_depth':[3,4,5,6,7,8,9,10],
              'criterion':['gini','entropy']}

gscv = GridSearchCV(estimator=dtc_gscv, param_grid=param_grid, scoring='accuracy', cv=3, n_jobs=-1, return_train_score=True)

scores_dtc_gscv = cross_val_score(gscv, d, t, cv=5)

print("Avr-Score-by-dtc_gscv:",scores_dtc_gscv.mean())
print("Std-Score-by-dtc_gscv:",scores_dtc_gscv.std())

gscv.fit(d,t)
```

```python
# cv_results_属性からrank_test_scoreとパラメータの組み合わせを取得
rank_test_score_dtc_gscv = gscv.cv_results_["rank_test_score"]
params_dtc_gscv = gscv.cv_results_["params"]

# スコアの良い順に並べ替えてハイパーパラメータを取得
sorted_params = [params_dtc_gscv[i] for i in np.argsort(rank_test_score_dtc_gscv)]
sorted_params
```

```python
gscv.cv_results_
```

```python
dtc_gscv_importances = gscv.best_estimator_.feature_importances_ 
#重要度を降順でソートした時に該当する元の配列のインデックスを表示する。ex)上位二つは元の配列における最も大きい数字と２番目に大きい数字のインデックスになっている。
index = np.argsort(dtc_gscv_importances)[::-1]


#特徴量を何個選択するか
for n in range(1,6):
  #行は丸ごと,列は重要度の高い特徴量に該当するもの
  new_d = d[:,index[:n]]
  new_scores_dtc = cross_val_score(gscv.best_estimator_, new_d, t, cv=5)
  print(n,'：avr：',new_scores_dtc.mean())
  print(n,'：std：',new_scores_dtc.std())

importance_name_std = [covtype.feature_names[i] for i in index[:8]]
importance_name_std
```

ランダムフォレスト

```python
rfc1 = RandomForestClassifier(n_estimators=30, max_depth=5, random_state=0,)
scores_rfc1 = cross_val_score(rfc1, d, t, cv=5)
print("Avr-Score-by-rfc1:",scores_rfc1.mean())
print("Std-Score-by-rfc1:",scores_rfc1.std())

rfc1.fit(d, t)
```

```python
rfc2 = RandomForestClassifier(n_estimators=30, random_state=0)

param_grid = {'max_depth':[3,4,5,6,7,8,9,10]}

gscv_rfc2 = GridSearchCV(estimator=rfc2, param_grid=param_grid, scoring='accuracy', cv=3, n_jobs=-1, return_train_score=True)

scores_gscv_rfc2 = cross_val_score(gscv_rfc2, d, t, cv=5)
print("Av-Score-by-gscv_rfc2:",scores_gscv_rfc2.mean())
print("Std-Score-by-gscv_rfc2:",scores_gscv_rfc2.std())

gscv_rfc2.fit(d, t)
```

```python
gscv_rfc2.cv_results_
```

```python
gscv_rfc2_importance = gscv_rfc2.best_estimator_.feature_importances_
gscv_rfc2_importance
```

```python
index_gscv_rfc2 = np.argsort(gscv_rfc2_importance)[::-1]

#ランダムフォレストにおけるmax_depthに関するグリッドサーチ結果の最適モデルの重要度により特徴量を40個に制限する
new_d_by_rfc2 = d[:, index_gscv_rfc2[:40]]

#これをbest_estimatorに適用
new_scores_rfc2 = cross_val_score(gscv_rfc2.best_estimator_, new_d_by_rfc2, t, cv=5)
print('：avr：',new_scores_rfc2.mean())
print('：std：',new_scores_rfc2.std())
```

```python
for n in range(40,48):
  new_d_by_rfc2 = d[:, index_gscv_rfc2[:n]]
  new_scores_rfc2 = cross_val_score(gscv_rfc2.best_estimator_, new_d_by_rfc2, t, cv=5)
  print(n,'：avr：',new_scores_rfc2.mean())
  print(n,'：std：',new_scores_rfc2.std())
  print(new_d_by_rfc2.shape)
```
